1) Problem formulation in Math:

Customer log behaviours from 2013 and 2014 are provided by Expedia. For each user event, the log contains 24 distinct attributes such as user id, check-in and check-out dates, hotel market, and origin to destination distance. These attributes can be seen as features describing each individual user event. We must generate probability distributions using this information for the likelyhood that a user event will be conducted in each of 100 hotel clusters. Using these probability distributions, we must classify each user event into a hotel cluster.


2) Math Solution:

Treating the customer log behaviour attributes as features, we will perform a Random Forest Classification. We believe this is a good classifier as Random Forests are known to deal with small samples (n) and large features (p) problems. Although there are hundreds of thousands of logs, there are 24 distinct attributes which makes classification relativly difficult using other classifiers. Random Forest Classification can generate probabilities for which we then can make predictions with. These predictions will then need to be evaluated using a metric.

```python
"""
Read/Collect Data
"""
# NOTE: Download Expedia Data from Kaggle, uncompress, and rename data paths to your data

# Read destinations, i.e. hotel clusters
destinations = pd.DataFrame.from_csv('C:\Users\Dennis\Desktop\input\destinations.csv')
print("Destinations converted to Pandas Data Frame")

# Read and sample train
train_rows = 0
with open('C:\Users\Dennis\Desktop\input\\train.csv') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    columns = spamreader.next()
    print('Read the train table columns')
    train_samp = []
    for i, row in enumerate(spamreader):
        if i % samp == 0:
            train_samp.append(row)
            train_rows += 1
            if train_rows == n_rows:
                break
        if i % 1e6 == 0:
            #print(i)
csvfile.close()
print("Train File Read")

# Build pandas Dataframe
train_samp = pd.DataFrame(train_samp, columns=columns)
train_samp.index = train_samp.user_id
train_samp.hotel_cluster = train_samp.hotel_cluster.astype(int)

# Merge Data (Default variable above is set to False)
if merge:
    train_samp = pd.merge(train_samp, destinations, left_on=train_samp.srch_destination_id.values.astype(int),
                          right_on=destinations.index.values, how='left')

# Export train file (Default variable above is set to None)
if train_file:
    train_samp.to_csv(train_file)

# Separate X_train and y_train
target = train_samp.hotel_cluster

# Removing excess columns in train
del train_samp['hotel_cluster']
del train_samp['cnt']
del train_samp['is_booking']

# Read test
test = pd.read_csv('C:\Users\Dennis\Desktop\input\\test.csv', index_col=0)
print("Test File Read")


"""
Preprocess Feature data
"""
del train_samp['user_id']
del test['user_id']

# Change NaN dates to '1970-01-01', data not used normally
train_samp[['date_time', 'srch_ci', 'srch_co']] = train_samp[['date_time',
                                                              'srch_ci', 'srch_co']].replace('', '1970-01-01')
test[['date_time', 'srch_ci', 'srch_co']] = test[['date_time', 'srch_ci', 'srch_co']].replace('', '1970-01-01')

# Remove NaNs
train_samp = train_samp.replace('', '9999')
test = test.replace('', '9999')
train_samp = train_samp.fillna(9999)
test = test.fillna(9999)

# Parse dates
def parse_dates(df): # takes data-frame and removes data info i.e. all 'string' data fields
    del df['date_time']
    del df['srch_ci']
    del df['srch_co']
    return df

train_samp = parse_dates(train_samp)
test = parse_dates(test)
print("Preprocessing done")
```
